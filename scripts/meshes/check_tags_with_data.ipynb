```python
import os
import glob
import pandas as pd
import numpy as np
```

```python
filtered_data = glob.glob("../../data/unprocessed/PS_LR_raw-all-data/*.obj")
actual_tags = pd.read_csv("../../data/unprocessed/PS_LR_all-tags/PS_LR_986-tags_Expert.csv")
# actual_tags = pd.read_csv("../../data/unprocessed/PS_LR_all-tags/PS_LR_986-tags_S3.csv")
```

```python
actual_tags_left = actual_tags[actual_tags["lateralidad"] == 0]
actual_tags_right = actual_tags[actual_tags["lateralidad"] == 1]
```

```python
actual_tags_left_ids = list(actual_tags_left["individuo"].unique())
actual_tags_right_ids = list(actual_tags_right["individuo"].unique())
```

```python
filtered_data_left_ids = []
filtered_data_right_ids = []
```

```python
filtered_data
```

```python
for datum in filtered_data:
    bone_id = int(datum.split("/")[-1].split("-")[0])
    bone_lt = datum.split("/")[-1].split("-")[1].split(".")[0][0].lower()

    if bone_lt == "l":
        filtered_data_left_ids.append(bone_id)
    elif bone_lt == "r":
        filtered_data_right_ids.append(bone_id)
    else:
        print(f"Problem with {datum}")
```

```python
missing_left = set(actual_tags_left_ids) - set(filtered_data_left_ids)
missing_left
```

```python
missing_right = set(actual_tags_right_ids) - set(filtered_data_right_ids)
missing_right
```

```python
tags_cleaned_l = actual_tags_left[~actual_tags_left["individuo"].isin(list(missing_left))]
tags_cleaned_l = tags_cleaned_l[tags_cleaned_l["individuo"] != 270]
```

```python
tags_cleaned_r = actual_tags_right[~actual_tags_right["individuo"].isin(list(missing_right))]
tags_cleaned_r = tags_cleaned_r[tags_cleaned_r["individuo"] != 270]
```

```python
tags_cleaned = pd.concat([tags_cleaned_l, tags_cleaned_r])
tags_cleaned
```

```python
# tags_cleaned.to_csv("../../data/unprocessed/PS_LR_all-tags/PS_LR_968-tags_Expert.csv", index=False)
tags_cleaned.to_csv("../../data/unprocessed/PS_LR_all-tags/PS_LR_968-tags_S3.csv", index=False)
```

```python
import pandas as pd
import os
import shutil
```

```python
in_path = "/media/valentino/Lothric/Panacea/datasets/shrec_16/"

train_df = pd.read_csv(in_path + "/train.csv", header=None)
test_df = pd.read_csv(in_path + "/test.csv", header=None)
val_df = pd.read_csv(in_path + "/val.csv", header=None)
```

```python
whole_df = pd.concat([train_df, test_df, val_df])
```

```python
whole_df = whole_df.rename(columns={0:"name", 1:"classes"})
```

```python
whole_df
```

```python
shrec_classes = sorted(whole_df["classes"].unique())
shrec_classes
```

```python
office_supply_1 = ["lamp", "laptop", "glasses"]
office_supply_2 = ["myScissor", "pliers", "cat"]
is_mammal = ["armadillo","camel", "cat", "centaur", "dog1", "dog2", "gorilla", "hand", "horse", "man", "rabbit", "santa", "woman"]
```

```python
out_path = "../../data/datasets/SHREC16"
```

```python
new_dataset = {"name" : [], "classes" : [], "classes_str" : [], "mammal" : [], "office" : []}
```

```python
for i in whole_df.iterrows():

    act_path = i[1].values[0]
    obj_name = os.path.split(act_path)[-1]
    # shutil.copy2("{}/{}".format(in_path, act_path),
    #              "{}/obj/{}".format(out_path, obj_name))
    new_dataset["name"].append(obj_name.split(".")[0])

    act_class = shrec_classes.index(i[1].values[1])

    is_mammal_act = 0
    if i[1].values[1] in is_mammal:
        is_mammal_act = 1

    is_office = 2
    if i[1].values[1] in office_supply_1:
        is_office = 0
    elif i[1].values[1] in office_supply_2:
        is_office = 1

    new_dataset["classes"].append(act_class)
    new_dataset["classes_str"].append(i[1].values[1])
    new_dataset["mammal"].append(is_mammal_act)
    new_dataset["office"].append(is_office)
```

```python
new_dataset
```

```python
new_dataset_df = pd.DataFrame(new_dataset)
```

```python
new_dataset_df.sort_values(by=["classes", "name"], inplace=True)
```

```python
new_dataset_df.to_csv("{}/dataset.csv".format(out_path), index=None)
```

```python
import os
import glob
```

```python
base_path = "../../data/unprocessed/PS_LR_raw-all-data"
```

```python
filtered_data = glob.glob("../../data/unprocessed/PS_LR_raw-all-data/*.obj")
```

```python
filtered_data
```

```python
for datum in filtered_data:
    bone_old_fname = os.path.split(datum)
    bone_id = int(datum.split("/")[-1].split("-")[0])
    bone_lt = datum.split("/")[-1].split("-")[1].split(".")[0][0].lower()

    if bone_lt == "i":
        os.rename(datum, os.path.join(base_path,f"{bone_id}-L.obj") )
    if bone_lt == "d":
        os.rename(datum, os.path.join(base_path,f"{bone_id}-R.obj") )
```

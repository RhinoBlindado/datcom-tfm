# Results 

```python
import pandas as pd
```

```python
results_df = pd.read_csv("../../docs/experiment_log__all.csv")
```

```python
results_df.columns
```

```python
# Filter out the "single" experiments from fn
single_res_df = results_df[results_df["fn"] == "single"]

# Now use the same for "multi" experiments
multi_res_df = results_df[results_df["fn"] != "single"]
```

# AF

```python
char = "af"

single_res_df[["exp_name", "size", "bone_type", f"{char}__test_f1", f"{char}__test_acc"]].sort_values(by=[f"{char}__test_f1"], ascending=False).head(10)
```

```python
multi_res_df[["exp_name", "size", "bone_type", "fn", f"{char}__test_f1", f"{char}__test_acc"]].sort_values(by=[f"{char}__test_f1"], ascending=False).head(30)
```
